# Notebook 00: FIA File Inventory and Table Selection

This notebook inventories the full FIA DataMart CSV folder without loading the large files into memory. It records file sizes, column counts and column names, then identifies the tables relevant to the project’s tree-level carbon storage prediction task.

The purpose is to support Section 2 of the coursework report: Data Understanding, Pre-processing and Exploration.

## 1. Import libraries and project paths

```python
import sys
from pathlib import Path
import pandas as pd

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.append(str(PROJECT_ROOT))

from src.paths import RAW_DIR, INVENTORY_DIR, create_project_dirs
from src.fia_inventory import create_file_inventory, create_table_columns_file

create_project_dirs()

print("Project root:", PROJECT_ROOT)
print("Raw data folder:", RAW_DIR)
print("Inventory folder:", INVENTORY_DIR)
```

## 2. Check raw FIA CSV files

```python
file_inventory_path = INVENTORY_DIR / "fia_file_inventory.csv"

file_inventory = create_file_inventory(
    raw_dir=RAW_DIR,
    output_path=file_inventory_path
)

file_inventory.sort_values("file_size_mb", ascending=False).head(20)
```

## 4. Create table-column inventory

```python
table_columns_path = INVENTORY_DIR / "fia_table_columns.csv"

table_columns = create_table_columns_file(
    raw_dir=RAW_DIR,
    output_path=table_columns_path
)

table_columns.head(20)
```

## 5. Inspect core tables for the carbon prediction project

```python
core_tables = [
    "ENTIRE_TREE.csv",
    "ENTIRE_PLOT.csv",
    "ENTIRE_COND.csv",
    "ENTIRE_REF_SPECIES.csv",
    "ENTIRE_REF_SPECIES_GROUP.csv",
    "ENTIRE_REF_FOREST_TYPE.csv",
    "ENTIRE_REF_FOREST_TYPE_GROUP.csv",
    "ENTIRE_COUNTY.csv",
    "ENTIRE_REF_OWNGRPCD.csv"
]

core_columns = table_columns[table_columns["file_name"].isin(core_tables)]

for table in core_tables:
    print("\n" + "="*80)
    print(table)
    print("="*80)
    print(core_columns.loc[core_columns["file_name"] == table, "column_name"].tolist())
```

## 6. Save table selection decisions

```python
selection_decisions = pd.DataFrame([
    {
        "file_name": "ENTIRE_TREE.csv",
        "include_for_project": "Yes",
        "reason": "Main individual-tree table containing tree measurements and carbon/biomass-related target variables."
    },
    {
        "file_name": "ENTIRE_PLOT.csv",
        "include_for_project": "Yes",
        "reason": "Provides plot-level location, inventory year and geographic context."
    },
    {
        "file_name": "ENTIRE_COND.csv",
        "include_for_project": "Yes",
        "reason": "Provides forest condition, stand, site and ownership variables."
    },
    {
        "file_name": "ENTIRE_REF_SPECIES.csv",
        "include_for_project": "Yes",
        "reason": "Provides species labels for interpreting SPCD."
    },
    {
        "file_name": "ENTIRE_REF_SPECIES_GROUP.csv",
        "include_for_project": "Yes",
        "reason": "Provides species group labels for business-friendly species analysis."
    },
    {
        "file_name": "ENTIRE_REF_FOREST_TYPE.csv",
        "include_for_project": "Yes",
        "reason": "Provides forest type labels for EDA and interpretation."
    },
    {
        "file_name": "ENTIRE_REF_FOREST_TYPE_GROUP.csv",
        "include_for_project": "Yes",
        "reason": "Provides broader forest type group labels."
    },
    {
        "file_name": "ENTIRE_COUNTY.csv",
        "include_for_project": "Yes",
        "reason": "Provides county and state labels for geographic analysis."
    },
    {
        "file_name": "ENTIRE_REF_OWNGRPCD.csv",
        "include_for_project": "Yes",
        "reason": "Provides ownership group labels for forest management interpretation."
    },
    {
        "file_name": "DWM, OZONE, LICHEN, VEG and FVS table families",
        "include_for_project": "No for first modelling dataset",
        "reason": "Useful for other FIA analyses, but not directly required for individual standing-tree carbon storage prediction."
    },
    {
        "file_name": "POP and EVALIDATOR table families",
        "include_for_project": "No for first modelling dataset",
        "reason": "Designed for population estimation and evaluation metadata rather than tree-level supervised regression."
    }
])

selection_path = INVENTORY_DIR / "fia_table_selection_decisions.csv"
selection_decisions.to_csv(selection_path, index=False)

selection_decisions
```

## 9. Distribution of above-ground carbon storage

```python
df["CARBON_AG"].describe(percentiles=[0.01, 0.05, 0.5, 0.95, 0.99]).to_frame()
```

```python
plot_carbon_distribution(df)
```

**Interpretation:** The raw above-ground carbon distribution is extremely right-skewed. Most trees have relatively low carbon storage, while a small number of very large trees hold exceptionally high carbon values. This pattern is important for the business problem because it suggests that carbon value is not evenly distributed across all trees. A small subset of larger trees may represent a disproportionate share of standing carbon, which is relevant for conservation prioritisation and sustainable harvesting review. From a modelling perspective, this skewness means that errors on high-carbon trees may be especially important and that a log transformation should be considered.

## 10. Log-transformed carbon distribution

```python
df["CARBON_AG_log"] = np.log1p(df["CARBON_AG"])

df["CARBON_AG_log"].describe().to_frame()
```

```python
plot_log_carbon_distribution(df)
```

**Interpretation:** The log-transformed carbon distribution is much more balanced than the raw target distribution, although it still shows multiple peaks. This suggests that the tree population contains several structural groups, such as small trees, medium-sized trees, and larger mature trees. The log transformation reduces the influence of extreme high-carbon trees and may help models such as Linear Regression perform more reliably. However, because the original business problem is about actual above-ground carbon storage, later modelling should clearly explain whether predictions are made on the raw carbon scale or transformed back from the log scale.

## 11. Relationship between diameter and carbon storage

```python
plot_dia_vs_carbon(eda_sample)
```

```python
df[["DIA", "CARBON_AG"]].corr()
```

**Interpretation:** Diameter at breast height shows a strong positive and clearly non-linear relationship with above-ground carbon storage. Carbon values increase slowly for small trees, then rise much more sharply as diameter increases. This confirms that `DIA` is likely to be one of the most important predictors in the modelling stage. The curved pattern also suggests that purely linear models may underfit larger trees unless engineered features such as `DIA_squared` or non-linear models such as Random Forest and Gradient Boosting are used. In business terms, this supports the idea that larger trees should receive particular attention in carbon-aware harvesting and conservation decisions.

## 12. Relationship between height and carbon storage

```python
plot_ht_vs_carbon(eda_sample)
```

```python
df[["HT", "ACTUALHT", "CARBON_AG"]].corr()
```

**Interpretation:** Tree height also has a positive relationship with above-ground carbon storage, but the relationship is more dispersed than the diameter-carbon relationship. Taller trees generally store more carbon, but trees of similar height can still vary widely in carbon storage. This suggests that height is useful, but it should be combined with diameter and species information rather than used alone. The correlation heatmap also shows that `HT` and `ACTUALHT` are almost perfectly correlated, so later preprocessing should consider whether both are needed or whether one should be dropped to reduce redundancy.

**Interpretation:** Crown ratio does not show a strong direct relationship with above-ground carbon storage. This suggests that `CR` is unlikely to be a primary predictor compared with diameter or height. However, it may still capture aspects of tree condition and crown structure, so it can be retained as a secondary predictor if missingness is manageable. Later model interpretation should confirm whether `CR` adds meaningful predictive value.

## 14. Carbon storage by diameter class

```python
df["diameter_class"] = pd.cut(
    df["DIA"],
    bins=[0, 5, 10, 20, 30, 50, 1000],
    labels=["0-5", "5-10", "10-20", "20-30", "30-50", "50+"]
)

diameter_class_summary = (
    df.groupby("diameter_class", observed=True)["CARBON_AG"]
    .agg(["count", "mean", "median"])
    .reset_index()
)

diameter_class_summary
```

```python
plot_carbon_by_diameter_class(df)
```

**Interpretation:** Median carbon storage increases sharply across diameter classes. The largest diameter group (`60+`) has a much higher median carbon value than all smaller groups, confirming that tree size is central to carbon storage. This chart is especially useful for communicating the business value of the model because diameter classes are easier for forest managers to interpret than individual scatter points. The result supports carbon-aware management decisions where larger trees may require closer review before harvesting because their removal could involve a higher carbon opportunity cost.

## 18. Correlation between numeric variables

```python
numeric_corr_cols = [
    "CARBON_AG",
    "DIA",
    "HT",
    "ACTUALHT",
    "CR",
    "STDAGE",
    "BALIVE",
    "ALSTK",
    "GSSTK"
]

numeric_corr_cols = [c for c in numeric_corr_cols if c in df.columns]

df[numeric_corr_cols].corr().round(3)
```

```python
plot_correlation_heatmap(df)
```

**Interpretation:** The correlation heatmap confirms that `DIA` has the strongest numeric relationship with above-ground carbon storage, with a correlation of about 0.83. `HT` and `ACTUALHT` also show strong positive relationships with carbon, both around 0.71, but they are almost perfectly correlated with each other at about 0.99. This creates a potential multicollinearity issue for linear models, so later preprocessing should consider using either `HT` or `ACTUALHT`, rather than both. `CR` has a weak correlation with carbon, while stand and stocking variables show weaker but potentially useful relationships. The negative relationship between `TPA_UNADJ` and tree size/carbon suggests it should be treated carefully and may not be suitable as a main predictor without further interpretation.

## 20. Summary

This notebook explored the cleaned FIA tree-level dataset prepared in Notebook 03. The EDA shows that above-ground carbon storage is highly right-skewed, meaning most trees store relatively low carbon while a smaller number of large trees account for much higher carbon values. This supports the business relevance of identifying high-carbon trees when making carbon-aware harvesting, conservation and regional planning decisions.

Tree diameter is the clearest predictor of above-ground carbon storage, with both the scatter plot and correlation heatmap showing a strong positive and non-linear relationship. Height also has a strong positive relationship with carbon, although `HT` and `ACTUALHT` are almost perfectly correlated, meaning the final preprocessing stage should avoid unnecessary duplication. Crown ratio appears much weaker as a direct predictor, but it may still be retained as a secondary condition-related feature if later modelling finds it useful.

The EDA also shows meaningful differences in median carbon storage across selected states, species codes and forest type codes. These patterns support the inclusion of geographic and ecological categorical variables in the modelling dataset. However, the current charts use FIA numeric codes, so the final report should translate key state, species and forest type codes into readable labels where possible.
